### Question 4: Write a Python program to find the average of all the elements in a list.


- `sample input : `
    - `var = [1,2,3,4,2,5]`
- `sample output :`
    - `result : 2.833 `

```python
# Please write The solution Here for above question 4:
var = [1, 2, 3, 4, 2, 5]

total_sum = 0
count = 0

for i in var:
    total_sum += i
    count += 1

avg = total_sum / count

print(f"result : {avg:.3f}")

```

### Question 7: Write a Python program to find which number is palindrome in below given list .

- `sample input : `
    - `var = [134,121,155]`
- `sample output :`
    - `result : 134 is not a palindome , 121 is a palindrome `

```python

# Please write The solution Here for above question 7:
var = [134, 121, 155]
results = []

for j in var:
    if str(j) == str(j)[::-1]:
        results.append(f"{j} is a palindrome")
    else:
        results.append(f"{j} is not a palindrome")
print("result : " + " , ".join(results))
```

### Question 8: Write a Python program to find the common elements between two lists.

- `sample input : `
    - `var1 = [1,2,3,4,10]`
    - `var2 = [10,11,12,14]`
- `sample output :`
    - `result : 10`

```python
# Please write The solution Here for above question 8:
var1 = [1, 2, 3, 4, 10]
var2 = [10, 11, 12, 14]
result = []

for i in var1:
    if i in var2:
        result.append(i)

print("result :", result)



```
